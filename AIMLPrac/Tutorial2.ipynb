## Series
1.pd.series(data=data ,index=index, name'something',dtype="")
2.pd.DataFrame(data,index, column)

```python
import numpy as np
import pandas as pd
```

```python
series = pd.Series(data={"a":1,"b":2,"c":3},name="Glue",dtype="int32")
print(series)
```

```python
series = pd.Series(data=(np.random.random(5)*10).astype('int32'),index=range(1,6),name="stick")
print(series)
```

```python
df = pd.DataFrame(data={"One":[1,1,1,1],"Two":[2,2,2,2]},index=range(1,5))
print(df)
```

```python
df=pd.DataFrame(data=[[1,2,3],[4,5,6],[7,8,9]],index=range(1,4))
df
```

```python
data ={
    "Science":(np.random.random(5)*100).astype('int32'),
    "Math":(np.random.random(5)*100).astype('int32'),
    "English":(np.random.random(5)*100).astype('int32')
}
index=["Adam","Bob","Charmander","Dave","Elon"]
df=pd.DataFrame(data=data,index=index)
df["Total%"]=round(df.sum(axis=1)/3,2)
df
```

```python
data =[
    {"a":1,"b":2},
    {"c":3,"d":4},
    {"a":5,"b":6}
]
pd.DataFrame(data, index=[1,2,3])
```
