# RNN DEMO

```python
import pandas as pd 
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
#Reading the dataset
df = pd.read_csv('spam.csv',encoding='ISO-8859-1')
df.head()
```

```python
x = df.iloc[:,1]
y = df.iloc[:,0]
```

```python
x.shape
```

```python
y.shape
```

```python
tokenizer =Tokenizer(oov_token='UNK')
```

```python
tokenizer.fit_on_texts(x)
```

```python
tokenizer.word_index
```

```python
seq = tokenizer.texts_to_sequences(x)
seq
```

```python
seq = pad_sequences(seq,padding='post')
seq
```

```python
seq.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(seq,y,test_size=0.4,random_state=42)
```

```python
X_train.shape ,X_test.shape,y_train.shape,y_test.shape
```

```python
X_train_red = pad_sequences(X_train,padding='post',maxlen=50)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers,models
from tensorflow.keras.layers import Embedding,LSTM,GRU,SimpleRNN,Dense,Dropout,Input
```

```python
model = models.Sequential(
    [
        Input(shape=(189,1)),#input layer
    SimpleRNN(16,activation='relu',return_sequences=False),#RNN layer,
    Dense(2,activation='softmax')#output layer
    ]
)
```

```python
model.summary()
```

```python
#Training
model.compile(loss='sparse_categorical_crossentropy',
optimizer='adam', metrics=['accuracy'])
X_train1,X_validation,y_train1,y_validation=train_test_split(X_train,y_train,random_state=42,test_size=0.2)
history = model.fit(X_train1,y_train1,epochs=5,validation_data=
(X_validation,y_validation))
```

```python
history_df = pd.DataFrame(history.history)
print(history_df.head())
```

```python
model.predict(X_test)
```

```python
#Evaluating
model.evaluate(X_test,y_test)
```
