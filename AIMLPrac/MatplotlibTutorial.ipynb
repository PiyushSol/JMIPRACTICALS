```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = np.linspace(-20,20,2000)
b1 = np.tanh(a)
b2 =np.sin(a)
b3 =np.cos(a)
b4=np.tan(a)
```

# Plotting the Graphs with grid of 2 and 2 matrix

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(4,4))
axes[0][0].plot(a,b1)
axes[0][1].plot(a,b2)
axes[1][0].plot(a,b3)
axes[1][1].plot(a,b4)
```

```python
axes[0][0].set_xlabel('Tan H')
axes[0][0].set_ylabel('Angle in radian')
```

```python
axes[0][0].plot(a,b1)
axes[0][1].plot(a,b2)
axes[1][0].plot(a,b3)
axes[1][1].plot(a,b4)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(4,4),dpi=100)
axes[0][0].plot(a,b1)
axes[0][1].plot(a,b2)
axes[1][0].plot(a,b3)
axes[1][1].plot(a,b4)
axes[0][0].set_xlabel('Tan H')
axes[0][0].set_ylabel('Angle in radian')
```

```python
fig.subplots_adjust(hspace=1,wspace=2)
```

## Plotting subplots in same graph , using legend to assign color to them

```python
fig,axes =plt.subplots(dpi=100)
```

```python
b1 = np.tanh(a)
b2 =np.sin(a)
b3 =np.cos(a)
b4=np.tan(a)
axes.plot(a,b1,label="tanh",color="#42e9f5",lw=0.5,ls="--")
axes.plot(a,b2,label="sin",color="#425af5",marker="v",ms="2.5")
axes.plot(a,b3,label="cos",color="#c242f5",lw=0.5,ls="--")
axes.legend()
plt.show()
```
