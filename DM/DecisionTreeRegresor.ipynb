```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../Datasets/AirQualityUCI.csv",sep=";",decimal=",")
```

```python
df.head()
```

```python
df_new = df.iloc[:,2:-2]
```

```python
df_new.head()
```

```python
df_new.isna().sum()
```

```python
df_new_category_cols = df_new.select_dtypes(include=['object','category'])
df_new_numerical_cols = df_new.select_dtypes(include=['int64','float64'])
```

```python
df_new_category_cols.fillna("Missing")
```

```python
df_new_numerical_cols.fillna(df_new_numerical_cols.mean(),inplace=True)
```

```python
df_modified = pd.DataFrame(df_new_numerical_cols)
```

```python
df_modified
```

```python
from sklearn.model_selection import train_test_split
X = df_modified.iloc[:,1:]
X.head()
```

```python
Y  =df_modified.iloc[:,0]
Y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,random_state=45)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_Scaled = scaler.fit_transform(X_train)
```

```python
X_test_Scaled = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model =DecisionTreeRegressor(criterion="squared_error",splitter="random",max_depth=5,random_state=45)
```

```python
model.fit(X_train_Scaled,y_train)
```

```python
y_predicted = model.predict(X_test_Scaled)
```

```python
y_predicted
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y_test,y_predicted)
```

```python
print(r2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
  
```

```python
plt.figure(figsize=(5,6),dpi=200)
sns.scatterplot(x=y_test,y=y_predicted,palette="deep")
sns.lineplot(x=y_test,y=y_test,color="blue",ls="--")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title(f"R2 Score Plot  (R2 = {r2 : 0.4f})")
plt.grid(True)
```
