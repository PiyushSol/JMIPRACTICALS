```python
#DecisionTreeClassification
import numpy as np 
import pandas as pd 
```

```python
#Reading the dataset
df = pd.read_csv("./Datasets/student-mat.csv",sep=";")
df.head()
```

```python
df.isna().sum()
```

```python
X = df.iloc[:, np.r_[0:21, 23:-1]]
X
```

```python
Y = df.iloc[:,22]
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder , StandardScaler
```

```python
X_train ,X_test , y_train,  y_test = train_test_split(X,Y,random_state=42)
```

```python
X_train.shape
```

```python

y_train.shape
```

```python
X_train_cols = X_train.select_dtypes(include='object')
X_train_num = X_train.select_dtypes(include='number')
```

```python
one = OneHotEncoder(drop="first",sparse_output=False)
scaler = StandardScaler()
```

```python
X_train_cols_enc = one.fit_transform(X_train_cols)
```

```python
X_train_cols_enc.shape
```

```python
X_train_num_enc = scaler.fit_transform(X_train_num)
X_train_num_enc
```

```python
X_test_cols = X_test.select_dtypes(include='object')
X_test_num = X_test.select_dtypes(include='number')
```

```python
X_test_cols_enc = one.transform(X_test_cols)
```

```python
X_test_num_enc = scaler.transform(X_test_num)
```

```python
df_cols_enc = pd.DataFrame(X_train_cols_enc)
df_num_enc = pd.DataFrame(X_train_num_enc)
X_train_encoded = pd.concat([df_cols_enc, df_num_enc], axis=1)
```

```python
df_test_cols_enc = pd.DataFrame(X_test_cols_enc)
df_test_num_enc = pd.DataFrame(X_test_num_enc)

X_test_encoded = pd.concat([df_test_cols_enc, df_test_num_enc], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
Y
```

```python
y_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=42)
```

```python
y_train_enc = le.fit_transform(y_train)
```

```python
y_test_enc = le.fit_transform(y_test)
```

```python
y_train_enc.shape
```

```python
model.fit(X_train_encoded,y_train_enc)
```

```python
y_pred = model.predict(X_test_encoded)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
cm = confusion_matrix(y_test_enc,y_pred)
```

```python
cm
```

```python
print("Classification Report : ")
print(classification_report(y_test_enc,y_pred))
```
